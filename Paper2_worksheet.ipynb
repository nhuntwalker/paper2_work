```python
def plot_radec_stats(df, counts, count_type=None, log=False):
    xmean = (df.ramin+df.ramax)/2.0
    ymean = (df.decmin+df.decmax)/2.0
    xbins = np.arange(0, 360+10., 10.)
    ybins = np.arange(-90, 100+10., 10.)
    
    H, xedges, yedges = binned_statistic_2d(xmean, ymean, counts, statistic="mean", bins=(xbins,ybins))
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

    ax = plt.subplot(1,1,1)
    if log==True:
        cax = ax.imshow(np.log10(H.T), extent=extent, aspect='auto', origin='lower')
    else:
        cax = ax.imshow(H.T, extent=extent, aspect='auto', origin='lower')
    if count_type != None:
        ax.text(0.9,0.1,count_type,transform=ax.transAxes, horizontalalignment="right")
    
    cbar = plt.colorbar(cax)    
    ax.set_ylim(-90,90)
    ax.set_xlabel('RA (deg)')
    ax.set_ylabel('Dec (deg)')
    ax.minorticks_on()
    plt.show()
```

```python
def plot_photometry_histo(df1, color, xaxis, xlabel, xlim=None, other_dfs=None, cutdata=True):
    fig = plt.figure(figsize=(9.3,4))
    fig.subplots_adjust(wspace=0.3)
    
    if type(cutdata) != type(None):
        cut1 = (df1.col12 < 1.5) & (df1.col12 > 1.0) & (df1.col23 < 3.5) & (df1.col23 > 2.5)
        df1 = df1[cut1]
    
    plt.subplot(121)
    if type(xlim) != type(None):
        xbins = np.linspace(xlim[0],xlim[1], 100)
    else:
        xbins = np.linspace(8,17,100)
        
    plt.hist(df1[xaxis], bins=xbins, log=True, histtype="step", linewidth=2, color=color)

    if type(other_dfs) != type(None):
        for the_set in other_dfs:
            df, col = the_set

            if type(cutdata) != type(None):
                cut = (df.col12 < 1.5) & (df.col12 > 1.0) & (df.col23 < 3.5) & (df.col23 > 2.5)
                df = df[cut]

            plt.hist(df[xaxis], bins=xbins, log=True, histtype="step", linewidth=2, color=col)
 
    else:
        med = np.median(df1[xaxis])
        plt.plot([med,med], [1,1E6], linestyle="--", color='k')

    if type(xlim) != type(None):
        plt.xlim(xlim[0],xlim[1])
    else:
        plt.xlim(8,16.2)

    plt.ylim(1)
    plt.xlabel(xlabel)
    plt.ylabel("N (count)")
    plt.minorticks_on()

    plt.subplot(122)
    H, xedges = np.histogram(df1[xaxis], bins=xbins)
    plt.plot(xedges[:-1], H/float(len(df1)), linewidth=2, color=color)

    if type(other_dfs) != type(None):
        for the_set in other_dfs:
            df, col = the_set
            if type(cutdata) != type(None):
                cut = (df.col12 < 1.5) & (df.col12 > 1.0) & (df.col23 < 3.5) & (df.col23 > 2.5)
                df = df[cut]

            H, xedges = np.histogram(df[xaxis], bins=xbins)
            plt.plot(xedges[:-1], H/float(len(df)), linewidth=2, color=col)
    else:
        med = np.median(df1[xaxis])
        plt.plot([med,med], [1,1E6], linestyle="--", color='k')

            
    if type(xlim) != type(None):
        plt.xlim(xlim[0],xlim[1])
    else:
        plt.xlim(8,16.2)

    plt.xlabel(xlabel)
    plt.ylabel("N (normed)")
    plt.minorticks_on()
    
    plt.show()
```

```python
def read_from_sql(fname, sep):
    df = pd.io.parsers.read_csv(fname, sep=sep, header=0)
    df = df[(df.cc_flags == '0000') & (df.w4mpro != -9999) & (abs(df.glat) > 40)]
    
    df['col12'] = df.w1mpro - df.w2mpro
    df['col23'] = df.w2mpro - df.w3mpro
    df['col34'] = df.w3mpro - df.w4mpro
    
    return df
```
